```python
import wave
import sys
import pyaudio

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100


def record_audio(seconds: int):
    output_path = "Customer.wav"
    with wave.open(output_path, "wb") as wf:
        p = pyaudio.PyAudio()
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)

        stream = p.open(format=FORMAT, channels=CHANNELS, rate=RATE, input=True)

        print("Recording...")
        for index in range(0, RATE // CHUNK * seconds):
            if index % (RATE // CHUNK) == 0:
                print(f"{index // (RATE // CHUNK)} / {seconds}s")
            wf.writeframes(stream.read(CHUNK))
        print("Done")

        stream.close()
        p.terminate()
    print(f"File saved at {output_path}")
    return output_path
```

```python
record_audio(20)
```

```python
from openai import OpenAI
client = OpenAI()

audio_file = open("Customer.wav", "rb")
transcript = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file, 
  response_format="text"
)
```

### Create embedding & search embeddings for right answer

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def get_embedding(text, model="text-embedding-ada-002"): # model = "deployment_name"
    return client.embeddings.create(input = [text], model=model).data[0].embedding

def search_docs(df, user_query, top_n=4, to_print=True):
    embedding = get_embedding(
        user_query,
        model="text-embedding-ada-002" # model should be set to the deployment name you chose when you deployed the text-embedding-ada-002 (Version 2) model
    )
    df["similarities"] = df.ada_v2.apply(lambda x: cosine_similarity(x, embedding))

    res = (
        df.sort_values("similarities", ascending=False)
        .head(top_n)
    )
    if to_print:
        display(res)
    return res


res = search_docs(Cust, transcript, top_n=1)
```

```python
print(res["Resolution"].values)
```

```python
context_embeddings=res["Resolution"].values
```

```python
user_query = transcript
```

### Call Generative model for final answer

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages =[
        {"role" : "system", "content" : "You are a helpful assistant."},
        {"role" : "user", "content" : f"User Query: \"{user_query}\"\nContext: \"{context_embeddings}\"\n\nAnswer: Given the user query \"{user_query}\" and the context \"{context_embeddings}\", generate a relevant and informative response."}
            ]
)
```
